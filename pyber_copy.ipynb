```python
%matplotlib inline
# Dependencies and Setup
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city = pd.read_csv(city_data_to_load)
ride = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride = pd.merge(city, ride, on = 'city', how = 'outer')

# Display the data table for preview
city_ride.head(10)
```

```python
# Gets the count per city grouped by type and each city.
driver_count_city = pd.DataFrame(city_ride.groupby(['type','city'])['driver_count'].mean())
driver_count_city
```

```python
city_types = list(set(city_ride['type']))

# Obtains dataframes for each city type

poop = [city_ride.loc[city_ride['type'] == citytype,:] for citytype in city_types]
    
num_rides = [poop.groupby('city')['ride_id'].count() for poop in poop]
avg_rides = [poop.groupby('city')['fare'].mean() for poop in poop]


# for x,y in zip(num_rides,avg_rides):
#     print(x.shape)
#     print(y.shape)
#     break

zero = 0

for x,y in zip(num_rides,avg_rides):
    for poop in range(zero,len(city_types)):
        color = input(f'Input color for {city_types[poop]} ')
        plt.scatter(x, y, s= 10*driver_count_city, c = color, edgecolors = 'black',alpha = 0.6)
        break
    zero +=1

plt.show()
            

```
